```python
from IPython.display import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig,(ax0,ax1) = plt.subplots(nrows=2, figsize=(9,6))
sigma = 1
mean = 0
x = mean + sigma * np.random.randn(10000)
# print(x)
ax0.hist(x,bins=40,density=False,histtype='bar',facecolor='yellowgreen',alpha=0.75)
ax1.hist(x,bins=20,density=1,histtype='bar',facecolor='pink',alpha=0.75,cumulative=True,rwidth=0.8)
plt.show()
```
